```python
import matplotlib.pyplot as plt
import pandas as pd
import io
import requests
import ipywidgets
from IPython.display import display
import plotly.graph_objects as go

#url="http://winterolympicsmedals.com/medals.csv"
url="https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23e1e9f0&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1168&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=BOGMBASE&scale=left&cosd=1959-01-01&coed=2020-01-01&line_color=%234572a7&link_values=false&line_style=solid&mark_type=none&mw=3&lw=2&ost=-99999&oet=99999&mma=0&fml=a&fq=Monthly&fam=avg&fgst=lin&fgsnd=2009-06-01&line_index=1&transformation=lin&vintage_date=2020-03-20&revision_date=2020-03-20&nd=1959-01-01"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))

df['DATE'] =  pd.to_datetime(df['DATE'], format='%Y-%m-%d')

df = df.sort_values('DATE', ascending=True)

#plt.plot(df['DATE'], df['BOGMBASE'])
#plt.xticks(rotation='vertical')


# Create figure
#fig = go.Figure()

#fig = go.FigureWidget()


#fig.add_trace(go.Scatter(x=list(df.DATE), y=list(df.BOGMBASE)))
trace=go.Scatter(x=list(df.DATE), y=list(df.BOGMBASE))
data = [trace]

layout = dict(
    title='Time series with range slider and selectors',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                    label='YTD',
                    step='year',
                    stepmode='todate'),
                dict(count=1,
                    label='1y',
                    step='year',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

def zoom(layout, xrange):
    in_view = df.loc[fig.layout.xaxis.range[0]:fig.layout.xaxis.range[1]]
    fig.layout.yaxis.range = [in_view.High.min() - 10, in_view.High.max() + 10]
    print("new")
    fig.show()
    


fig = go.Figure(data=data, layout=layout)
fig.layout.on_change(zoom, 'xaxis.range')


fig
```

```python
import plotly.graph_objects as go


import numpy as np
np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)

f = go.FigureWidget([go.Scatter(x=x, y=y, mode='markers')])

scatter = f.data[0]
colors = ['#a3a7e4'] * 100
scatter.marker.color = colors
scatter.marker.size = [10] * 100
f.layout.hovermode = 'closest'



# create our callback function
def update_point(trace, points, selector):
    print("d")
    c = list(scatter.marker.color)
    s = list(scatter.marker.size)
    for i in points.point_inds:
        c[i] = '#bae2be'
        s[i] = 20
        with f.batch_update():
            scatter.marker.color = c
            scatter.marker.size = s

scatter.on_click(update_point)

f
#f.show()
```
